# Code to check credit limit

change the file path in pd.read_excel()

```python
import pandas as pd

# Load the responses
responses = pd.read_excel("feedback_form_responses.xlsx")

# Initialize an empty list to store the indices of rows to keep
rows_to_keep = []

emails = responses['Email Address']
choices = list(responses['Courses that I plan to take in the upcoming semester'])

# Loop through each person's course choices
for index, (email, person_choice) in enumerate(zip(emails, choices)):
    person_choice = [x.strip() for x in person_choice.split(',')]
    
    # Remove specific courses
    if "BIO339/639" in person_choice:
        person_choice.remove("BIO339/639")
    if "BIO416/716" in person_choice:
        person_choice.remove("BIO416/716")
    
    # Calculate total credits
    credits = 0
    for course in person_choice:
        credits += int(course[-1])  # Assuming the credit is the last character of the course string
    
    # If total credits are less than or equal to 25, mark the row to keep
    if credits <= 25:
        rows_to_keep.append(index)

# Filter the DataFrame to keep only the rows with total credits <= 25
filtered_responses = responses.iloc[rows_to_keep]

# Save the filtered DataFrame to a new Excel file
filtered_responses.to_excel("filtered_feedback_form_responses.xlsx", index=False)

filtered_responses 
```

# Code to check how many students per course

```python
responses = pd.read_excel("feedback_form_responses.xlsx")
emails = responses['Email Address']
choices = list(responses['Courses that I plan to take in the upcoming semester'])
subjects = []

for person_choice in choices:
    person_choice = [x.strip() for x in person_choice.split(',')]
    if "BIO339/639" in person_choice:
        person_choice.remove("BIO339/639")
    if "BIO416/716" in person_choice:
        person_choice.remove("BIO416/716")
    subjects.extend(person_choice)

subjects = set(subjects)

subject_count = {}
for subject in subjects:
    count = 0
    if "MTH" not in subject:
        pass
    for person_choice in choices:
        if subject in person_choice:
            count += 1
    subject_count[subject] = count

subject_count
```

```python
course_interelations = {}

for subject in subject_count:
    common_courses = {}
    
    for other_subject in subject_count:
        if subject != other_subject:
            common_courses[other_subject] = 0
            for person_choice in choices:
                if subject in person_choice and other_subject in person_choice:
                    common_courses[other_subject] += 1
    course_interelations[subject] = common_courses


pprint.pprint(course_interelations)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert the dictionary to a DataFrame
df = pd.DataFrame(course_interelations)

# Fill NaN values with 0 (if any)
df = df.fillna(0)

# Generate the heatmap
plt.figure(figsize=(30, 30))
heatmap = sns.heatmap(df, annot=True, cmap='YlGnBu', cbar=True, linewidths=.5)

# Set titles and labels
plt.title('Course Interrelations Heatmap')
plt.xlabel('Courses')
plt.ylabel('Courses')

# Show the heatmap
plt.show()

# To save the heatmap as an image
plt.savefig('course_interrelations_heatmap.png')
```
